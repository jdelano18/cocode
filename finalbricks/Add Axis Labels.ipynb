## Create example mathematical function

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
```

## Plot it

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.show()
```
