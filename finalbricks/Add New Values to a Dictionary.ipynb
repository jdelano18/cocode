```python
# adds element if key doesn't exist in dict
# if key exists in dict, update existing value
dictionary.update({key : value})
```

```python
# does the same thing
dictionary[key] = value
```

```python
# inserts key:value pair *only* if key doesn't exist in dict
dictionary.setdefault(key, value)
```

## Using update

```python
car = {'make' : 'tesla', 'model' : 'model 3', 'year' : 2019}
car.update({'model' : 'model s'})
car.update({'color' : 'space grey'})
car
```

## Classic assignment syntax

```python
# this does the same thing as the cell above
# just use whatever syntax you feel more comfortable with
car = {'make' : 'tesla', 'model' : 'model 3', 'year' : 2019}
car['model'] = 'model s'
car['color'] = 'space grey'
car
```

## Using setdefault

```python
car = {'make' : 'tesla', 'model' : 'model 3', 'year' : 2019}
car.setdefault('model', 'model s') # ** won't change anything **
car.setdefault('color', 'space grey')
car
```
