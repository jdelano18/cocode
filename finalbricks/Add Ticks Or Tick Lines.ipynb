        <p>The axes have little marks called ticks. To be precise, ticks are the locations of the marks (eg. (-1, 0, 1)), tick lines are the small lines drawn at those locations, tick labels are the labels drawn next to the tick lines, and tickers are objects that are capable of deciding where to place ticks. The default tickers typically do a pretty good job at placing ~5 to 8 ticks at a reasonable distance from one another.</p>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
ax.set_xticks(np.arange(-2, 2, 1))
ax.set_yticks([-2, 0, 1, 2])
```

```python
# activate minor ticks
plt.minorticks_on()
```

```python
import numpy as np
x = np.linspace(-2, 2, 100)

fig, ax = plt.subplots()
plt.plot(x, x**3)
ax.set_xticks(np.arange(-2, 2, 1))
plt.grid(True)
plt.title("Manual ticks on the x-axis")
plt.show()
```

```python
fig, ax = plt.subplots()
plt.plot(x, x**3)
plt.minorticks_on()
ax.set_xticks([-2, 0, 1, 2])
ax.set_yticks(np.arange(-5, 5, 1))
ax.yaxis.set_ticklabels(["min", -4, -3, -2, -1, 0, 1, 2, 3, "max"])
plt.title("Manual ticks and tick labels\n(plus minor ticks) on the y-axis")
plt.grid(True)
plt.show()
```
