## Make File With Some Content

```
echo "Jan" >> sales.txt; echo "Feb" >> sales.txt; echo "Mar" >> sales.txt
```

## Make File Contents

```
cat sales.txt
```
```
Jan
Feb
Mar
```

## Make File Contents With Line Numbers

```
cat -n sales.txt
```
```
     1	Jan
     2	Feb
     3	Mar
```
