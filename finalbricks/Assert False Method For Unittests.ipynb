```python
def is_even(n):
    return n % 2 == 0
    
class Tester(unittest.TestCase):
    def test_is_even(self):
        n = is_even(19)
        self.assertFalse(n)

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
