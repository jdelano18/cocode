```python
def my_func(x):
    return abs(x) * 10
    
class Tester(unittest.TestCase):
    def test_not_negative(self):
        n = my_func(-134)
        self.assertGreater(n, -1)

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
