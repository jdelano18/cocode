```python
import http.client as httplib
def isonline():
    """
    Check if you are currently connected to the internet.
    """

    conn = httplib.HTTPConnection("www.cocode.ai", timeout=5)
    try:
        conn.request("HEAD", "/")
        conn.close()
        return True
    except:
        conn.close()
        return False
```
