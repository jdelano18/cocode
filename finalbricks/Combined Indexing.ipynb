```python
import numpy as np
```

```python
X = np.arange(12).reshape((3, 4))
X
```

```python
# fancy and simple indexing
X[2, [2, 0, 1]]
```

```python
# fancy indexing and slices 
X[1:, [2, 0, 1]]
```

```python
# fancy indexing and masking
row = np.array([0, 1, 2])
mask = np.array([1, 0, 1, 0], dtype=bool)
X[row[:, np.newaxis], mask]
```
