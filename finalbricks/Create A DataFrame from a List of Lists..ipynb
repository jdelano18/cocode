## Create a list of lists that is our data

```python
values = [
            [1985, 1, "Biking", 68],
            [1984, 3, "Dancing", 83],
            [1992, 0, "Skiing", 112]
         ]
```

## Create DataFrame from `values` with column names and index labels

```python
d3 = pd.DataFrame(
        values,
        columns=["birthyear", "children", "hobby", "weight"],
        index=["alice", "bob", "charles"]
     )
d3
```
