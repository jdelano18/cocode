```python
# Example&# mapping a list of first names to last names
firstnames = ['John', 'Sally','Iris']
lastnames = ['Smith', 'Brown', 'Johnson']
my_dict = {first:last for (first,last) in zip(firstnames, lastnames)}
my_dict
```
