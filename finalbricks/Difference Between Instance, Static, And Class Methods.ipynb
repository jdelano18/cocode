## Instance Methods

Instance methods must take `self` as the first parameter. These are the most common type that you'll use most of the time.

```python
class Human:
    species = "H. sapiens"
    
    def __init__(self, name):
        self.name = name
        self._age = 0
    
    def hi(self):
        print(f"hi my name is {self.name}")
```

```python
me = Human("Jimmy")
me.hi()
```

## Class Methods

Class methods are bound to class itself not the instance of any particular object. 

```python
class Human:
    species = "H. sapiens"

    def __init__(self, name):
        self.name = name
        self._age = 0

    # A class method is shared among all instances
    @classmethod
    def get_species(cls):
        return cls.species
```

```python
bob = Human("bob")
bob.get_species()
```

## Static Methods

```python
class Human:
    species = "H. sapiens"

    def __init__(self, name):
        self.name = name
        self._age = 0
    
    @staticmethod
    def speak():
        return "yo "*3
```

```python
alice = Human("alice")
alice.speak()
```
