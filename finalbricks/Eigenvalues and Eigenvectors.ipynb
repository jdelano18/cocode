```python
# compute λ, v for square matrix m
eigenvalues, eigenvectors = linalg.eig(m)
```

```python
m3 = np.array([[1,2,3],[5,7,11],[21,29,31]])
m3
```

```python
eigenvalues, eigenvectors = linalg.eig(m3)
eigenvalues # λ
```

```python
eigenvectors # v
```

```python
m3.dot(eigenvectors) - eigenvalues * eigenvectors  # m3.v - λ*v = 0
```
