```python
def difference(num1, num2):
    """
    Find the difference between 2 numbers.

    :type num1: number
    :param num1: The first number to use.

    :type num2: number
    :param num2: The second number to use.

    >>> difference(1, 4)
    3
    """

    # Return the calculated value
    return abs(num1 - num2)


```
