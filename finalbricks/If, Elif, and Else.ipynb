# If, Elif, and Else

if, elif, and else statements are used for decision making in python.
## ask about picture

## Example

```python
# Let's just make a variable
some_var = 5
```

```python
if some_var > 10:
    print("some_var is totally bigger than 10.")
elif some_var < 10:
    print("some_var is smaller than 10.")
else:
    print("some_var is indeed 10.")
```
