```python
import pandas as pd
```

```python
data = {('California', 2000): 33871648,
        ('California', 2010): 37253956,
        ('Texas', 2000): 20851820,
        ('Texas', 2010): 25145561,
        ('New York', 2000): 18976457,
        ('New York', 2010): 19378102}
pop = pd.Series(data)
pop
```

```python
# single element
pop['California', 2000]
```

```python
# partial indexing one level
pop['California']
```

```python
# partial slicing (if indicies are sorted)
pop = pop.sort_index()
pop.loc['California':'New York'] 
```

```python
# partial indexing on lower levels
pop[:, 2000]
```

```python
# boolean masks
pop[pop > 22000000]
```

```python
# fancy indexing
pop[['California', 'Texas']]
```
