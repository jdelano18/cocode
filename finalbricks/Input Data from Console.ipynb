## Prompting for a Value

```python
# try running it for yourself!
name = input("What's your name? ")
print("Hello, " + name + "!")
```

## Prompting for a numerical input

```python
age = input("How old are you? ")
age = int(age)
print(age)
```

```python
pi = input("What's the value of pi? ")
pi = float(pi)
print(pi)
```
