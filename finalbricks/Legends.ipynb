```python
import matplotlib.pyplot as plt
```

```python
plt.legend(loc='location')
```

The location could be something like 'upper right' or 'lower left'. The string 'best' places the legend at the location with the minimum overlap with other stuff. This option can be quite slow for plots with large amounts of data -- instead, you may find faster performance by providing a specific location.

See more in [the documentation](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.legend.html)
