# List Slicing

## Look at ranges of a list with slice syntax

The start index is included, the end index is not. (It's a closed/open range for you mathy types.)

````python
l = [1,5,7,9,11,13,15,17,19]
l[1:3]
````

## Use any combination of ```your_list[start:end:step]```

````python
# Omit the beginning and return the list
l[2:]
````

````python
# Omit the end and return the list
l[:3]
````

````python
# Select every second entry
l[::2]
````

````python
# Return a reversed copy of the list
l[::-1]
````
