# Lists Basics in Python

## Lists store sequences of objects in python. They don't  have to be filled with objects of the same type 

````python
a_list = [3, "hey", [(2,10)], {'woah': 21}]
a_list
````

## Add an element to the end of a list with append

````python
l = []
l.append(1)    # l is now [1]
l.append(2)    # l is now [1, 2]
l.append(4)    # l is now [1, 2, 4]
l.append(3)
l
````

## Remove from the end with pop

````python
l.pop()        # => 3 and l is now [1, 2, 4]
````

## Access a list like you would any array

````python
l[0]
````

````python
# Look at the last element
l[-1]  # => 4
````

````python
# Looking out of bounds is an IndexError
l[10]  # Raises an IndexError
````

## Copy an array

````python
# Make a one layer deep copy using slices
l2 = l[:]  # => l2 = [1, 2, 4] but (l2 is l) will result in false.
````

## Removing elements with ```del``` and ```remove()```

````python
# remove a particular element by index with del
del l[2]  # li is now [1, 2]
````

````python
# Remove first occurrence of a value
l.remove(1)  # li is now [2]
l.remove(1)  # Raises a ValueError as 1 is not in the list
````

## Index method

````python
# Get the index of the first item found matching the argument
l.index(2)  # => 0
````

## Insert method

````python
# Insert an element at a specific index
l.insert(1, 1)
l.insert(2,"hey")
l
````

## Adding lists

````python
l + [7,8,9]
````

## Check for existence with ```in```

````python
"hey" in l
````

## Length of list with ```len()```

````python
len([1,3,5,7,9])
````
