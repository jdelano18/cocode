```python
import matplotlib.pyplot as plt
```

```python
plt.yscale('logit')
```

```python
import numpy as np
x = np.linspace(0.1, 15, 500)
y = x**3/np.exp(2*x)

plt.figure(3)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
plt.show()
```
