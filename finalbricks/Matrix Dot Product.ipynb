## Create two matrices 

```python
n1 = np.arange(10).reshape(2, 5)
n1
```

```python
n2 = np.arange(15).reshape(5,3)
n2
```

## Compute a matrix dot product using the `dot` method

```python
n1.dot(n2)
```
