## Create a matrix

```python
import numpy.linalg as linalg

m3 = np.array([[1,2,3],[5,7,11],[21,29,31]])
m3
```

## Calculate its inverse

```python
linalg.inv(m3)
```

## Compute the [pseudoinverse](https://en.wikipedia.org/wiki/Moore%E2%80%93Penrose_pseudoinverse) using `pinv`:

```python
linalg.pinv(m3)
```
