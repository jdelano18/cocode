```python
import numpy as np
```

## Create arrays

```python
x = np.array([7, 2, 3, 1, 6, 5, 4])
x
```

```python
rand = np.random.RandomState(42)
M = rand.randint(0, 10, (4, 6))
print(M)
```

## Partition

```python
np.partition(x, 3)
```

```python
np.partition(M, 2, axis=1)
```
