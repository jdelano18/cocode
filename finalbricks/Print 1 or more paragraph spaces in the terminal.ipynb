```python
def newline(lines=1):
    """
    Print 1 or more paragraph spaces in the terminal.

    :type lines: integer
    :param lines: The amount of paragraph spaces to print.
    """

    # Print the new line iterated by the amount of new lines
    print('\n' * lines)
```
