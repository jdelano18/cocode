```python
class Example:
    def __init__(self, x):
        self._x = x
    
    @property
    def x(self):
        return self._x
```

```python
class Human:
    species = "H. sapiens"

    def __init__(self, name):
        self._name = name
        self._age = 0
    
    @property
    def name(self):
        return self._name
    
    @property
    def age(self):
        return self._age
```

```python
alice = Human("alice")
print(alice.name)
print(alice.age)
```
