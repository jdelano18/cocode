```python
class Example:
    def __init__(self, x):
        self._x = x
    
    @property
    def x(self):
        return self._x
    
    @x.setter
    def x(self, x):
        self._x = x
```

```python
class Human:
    species = "H. sapiens"

    def __init__(self, name):
        self._name = name
        self._age = 0
    
    @property
    def age(self):
        return self._age
    
    @age.setter
    def age(self, age):
        self._age = age
```

```python
alice = Human("alice")
print(alice.age)
alice.age = 23
print(alice.age)
```
