```python
a = np.array([14, 23, 32, 41])
b = np.array([5,  4,  3,  2])
print("a - b  =", a - b)
```

The `-` operator implicitly calls the `.subtract()` ufunc

```python
print("a + b  =", np.subtract(a,b))
```
