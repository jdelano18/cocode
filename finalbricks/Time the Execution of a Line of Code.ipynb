```python
# time a 1-liner
%time sum(range(100))
```

```python
%timeit sum(range(100))
```

```python
# use %% for multi-line code
%%timeit
total = 0
for i in range(1000):
    for j in range(1000):
        total += i * (-1) ** j
```
