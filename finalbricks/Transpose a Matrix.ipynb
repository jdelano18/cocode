## The `T` attribute has no effect on rank 0 (empty) or rank 1 arrays:

```python
m2 = np.arange(5)
m2
```

```python
m2.T
```

## We can get the desired transposition by first reshaping the 1D array to a single-row matrix (2D):

```python
m2r = m2.reshape(1,5)
m2r
```

```python
m2r.T
```
