# Variables and Assignment

### There are no declarations, only assignments. Convention is to use lower_case_with_underscores

````python
some_var = 5
some_var
````

### ```if``` can be used in a variable assignment

````python
a = "hey!" if 3 > 2 else "hello"
a
````
