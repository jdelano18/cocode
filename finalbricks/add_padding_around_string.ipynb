## Create Some Text 

```python
text = 'Chapter 1'
```

## Add Padding Around Text 

```python
# Add Spaces Of Padding To The Left
format(text, '>20')
```

```python
# Add Spaces Of Padding To The Right
format(text, '<20')
```

```python
# Add Spaces Of Padding On Each Side
format(text, '^20')
```

```python
# Add * Of Padding On Each Side
format(text, '*^20')
```
