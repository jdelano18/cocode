## Change To Integer

```scala
// Convert a float to an integer 
32.34.toInt
```

```scala
// Convert a string to an integer
"23394".toInt
```

## Change To Double

```scala
// Convert a string to a double
"23394".toDouble
```

## Change To Float

```scala
// Convert a string to float
"23394".toFloat
```

```scala
// Convert an integer to float
3923.toFloat
```

## Change To Long

```scala
// Convert a string to long
"23394".toLong
```

## Change To Short

```scala
// Convert a string to short
"23394".toShort
```

## Change To Byte

```scala
// Convert a string to byte
"10".toByte
```

## Change To String

```scala
// Convert an integer to string
2393.toString
```
