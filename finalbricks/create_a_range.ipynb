## Create A Range 1 to 10

```scala
// Create a range between 1 and 10
1 to 10
```

## Create A Range In An Array

```scala
// Create an array between 1 and 10 and put in an array
(1 to 10).toArray
```

## Use A Range In A For Loop

```scala
// For each 1 in 1,2,3,4,5
for (i <- 1 to 10) 
    // Print i
    println("index: "+ i)
```
