```python
# Import modules
from datetime import datetime
from datetime import timedelta
```

```python
# Create a variable with the current time
now = datetime.now()
now
```

```python
# The current year
now.year
```

```python
# The current month
now.month
```

```python
# The current day
now.day
```

```python
# The current hour
now.hour
```

```python
# The current minute
now.minute
```

```python
# The difference between two dates
delta = datetime(2011, 1, 7) - datetime(2011, 1, 6)
delta
```

```python
# The difference days
delta.days
```

```python
# The difference seconds
delta.seconds
```

```python
# Create a time
start = datetime(2011, 1, 7)
```

```python
# Add twelve days to the time
start + timedelta(12)
```
