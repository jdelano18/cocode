## Preliminaries

```python
# Load library
import pandas as pd
```

## Create Feature Matrix

```python
# Create features
df = pd.DataFrame({'Score': ['Low', 
                             'Low', 
                             'Medium', 
                             'Medium', 
                             'High']})

# View data frame
df
```

## Create Scale Map

```python
# Create mapper
scale_mapper = {'Low':1, 
                'Medium':2,
                'High':3}
```

## Map Scale To Features

```python
# Map feature values to scale
df['Scale'] = df['Score'].replace(scale_mapper)

# View data frame
df
```
