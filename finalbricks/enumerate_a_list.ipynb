```python
# Create a list of strings
data = ['One','Two','Three','Four','Five']
```

```python
# For each item in the enumerated variable, data
for item in enumerate(data):
    # Print the whole enumerated element
    print(item)
    # Print only the value (not the index number)
    print(item[1])
```
