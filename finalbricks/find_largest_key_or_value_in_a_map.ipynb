## Create A Map

```scala
// Create an immutable map with three key value pairs
val numbers = Map(1 -> 100, 
                  2 -> 200,
                  3 -> 300)
```

## Find Largest Key

```scala
// Find largest key
numbers.max
```

## Find Largest Value

```scala
// Find the largest value
numbers.valuesIterator.max
```
