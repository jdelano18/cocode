## Create Two Lists

```python
names = ['James', 'Bob', 'Sarah', 'Marco', 'Nancy', 'Sally']
ages = [42, 13, 14, 25, 63, 23]
```

## Iterate Over Both Lists At Once

```python
for name, age in zip(names, ages):
     print(name, age)
```
