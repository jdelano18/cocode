### Import the math module

```python
import math
```

### Display the value of pi.

```python
math.pi
```

### Display the value of e.

```python
math.e
```

### Sine, cosine, and tangent

```python
math.sin(2 * math.pi / 180)
```

### Exponent

```python
2 ** 4, pow(2, 4)
```

### Absolute value

```python
abs(-20)
```

### Summation

```python
sum((1, 2, 3, 4))
```

### Minimum

```python
min(3, 9, 10, 12)
```

### Maximum

```python
max(3, 5, 10, 15)
```

### Floor

```python
math.floor(2.949)
```

### Truncate (drop decimal digits)

```python
math.trunc(32.09292)
```

### Truncate (integer conversion)

```python
int(3.292838)
```

### Round to an integrer

```python
round(2.943), round(2.499)
```
