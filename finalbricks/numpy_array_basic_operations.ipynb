```python
# Import modules
import numpy as np
```

```python
# Create an array
civilian_deaths = np.array([4352, 233, 3245, 256, 2394])
civilian_deaths
```

```python
# Mean value of the array
civilian_deaths.mean()
```

```python
# Total amount of deaths
civilian_deaths.sum()
```

```python
# Smallest value in the array
civilian_deaths.min()
```

```python
# Largest value in the array
civilian_deaths.max()
```
