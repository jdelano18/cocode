## Preliminaries

```python
import pandas as pd
```

## Create Dataframe

```python
# Create empty dataframe
df = pd.DataFrame()

# Create a column
df['name'] = ['John', 'Steve', 'Sarah']

# View dataframe
df
```

## Assign New Column To Dataframe

```python
# Assign a new column to df called 'age' with a list of ages
df.assign(age = [31, 32, 19])
```
