```python
import pandas as pd
```

```python
# Create an example dataframe
data = {'name': ['Jason', 'Molly'], 
        'country': [['Syria', 'Lebanon'],['Spain', 'Morocco']]}
df = pd.DataFrame(data)
df
```

```python
df[df['country'].map(lambda country: 'Syria' in country)]
```
