## Preliminaries

```python
import pandas as pd
%matplotlib inline
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame()

df['x'] = random.sample(range(1, 100), 25)
df['y'] = random.sample(range(1, 100), 25)
```

```python
df.head()
```

## Scatterplot

```python
sns.lmplot('x', 'y', data=df, fit_reg=False)
```

## Density Plot

```python
sns.kdeplot(df.y)
```

```python
sns.kdeplot(df.y, df.x)
```

```python
sns.distplot(df.x)
```

## Histogram

```python
plt.hist(df.x, alpha=.3)
sns.rugplot(df.x);
```

## Boxplot 

```python
sns.boxplot([df.y, df.x])
```

## Violin Plot

```python
sns.violinplot([df.y, df.x])
```

## Heatmap

```python
sns.heatmap([df.y, df.x], annot=True, fmt="d")
```

## Clustermap

```python
sns.clustermap(df)
```
