### Setup the originally variables and their values

```python
one = 1
two = 2
```

### View the original variables

```python
'one =', one, 'two =', two
```

### Swap the values

```python
one, two = two, one
```

### View the swapped values, notice how the values for each variable have changed

```python
'one =', one, 'two =', two
```
