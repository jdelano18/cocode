## Preliminaries

```python
from collections import namedtuple
```

## Create A Named Tuple

```python
Vehicle = namedtuple('Vehicle', 'make model wheels manual')
```

## Create An Entry

```python
forrester = Vehicle('Forrester', 'Subaru', 4, True)
```

## View The Data In Entry

```python
forrester.model
```

```python
forrester.wheels
```
