# Key Code&

If no value is passed for the default_value, var will equal the default_value. Otherwise, var will equal whatever value you pass into the function.

```python
def my_function(var = default_value):
    pass
```

# Example&

```python
def my_function(name = "mom"):
    print(f"hi {name}")
    
my_function()
my_function("grandma")
```
