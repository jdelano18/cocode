The ```append()``` method adds an object to the end of the list.

The ```extend(iterable_object)``` method extends an object by appending each element in the iterable object.

---

# Example&

````python
l = [1,2,3]
l.append([4,5])
l
````

````python
l = [1,2,3]
l.extend([4,5])
l
````

````python
l.extend(6)
````
