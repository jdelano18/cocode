Using the ```pass``` statement in an else block allows you to fail silently.

# Example&

````python
f_names = ['alice.txt', 'siddhartha.txt','moby_dick.txt', 'little_women.txt']
for f_name in f_names:
 # Note: I just made up random numbers for the number of lines
 #       These are not accurate unless I was really lucky
    try:
        with open(f_name) as f_obj:
            lines = f_obj.readlines()
    except FileNotFoundError:
            # Just move on to the next file.
        pass
    else:
        num_lines = len(lines)
        msg = f"{f_name} has {num_lines} lines."
        print(msg)
````
