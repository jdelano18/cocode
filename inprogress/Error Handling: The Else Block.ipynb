The ```else``` block should contain any code that depends on the ```try``` block running successfully. 

# Example&

````python
def try_to_divide(num, denom):
    try:
        result = int(num) / int(denom)
    except ZeroDivisionError:
        print("You can't divide by zero!")
    else:
        print(result)
````

````python
try_to_divide(6, 2)
````

````python
try_to_divide(6, 0)
````
