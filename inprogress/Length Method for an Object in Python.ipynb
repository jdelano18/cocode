# Key Code&

Many python objects (like lists, sets, strings, tuples, and more) have a built-in ```__len__()``` method that can be called by ```len(object)```

````python
print(len("this string") == "this string".__len__())
print(len(["of", "this", "list", "of", "strings"]))
````
