# Key Code&

```python
# iterable from lower up to but not including upper
# by default lower = 0 and step = 1
range(lower, upper, step)
```

# Example&

```python
for i in range(4):
    print(i)
```

# Example&

```python
for i in range(4, 8):
    print(i)
```

# Example&

```python
# If step is not indicated, the default value is 1.
for i in range(4, 10, 2):
    print(i)
```

# Example&

## Using a negative number for the step will reverse the numbers

```python
for i in range(5,1, -1):
    print(i)
```
