# Key Code&

```python
some_set ^ other_set ^ another_one...
```

# Example&

```python
primes = {2,3,5,7,11,13}
odds = {1,3,5,7,9,11,13,15}
pow2 = {2,4,8,16}
odds ^ primes ^ pow2
```
